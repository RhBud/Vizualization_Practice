```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred')
```

```python
plt.hist(dataset1,normed=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
#normed=True controls for the fact that both datasets have different numbers of data entries
#alpha= gives option for setting transparancy levels
#bins= gives us option to select number of bins other than the default 10
```
