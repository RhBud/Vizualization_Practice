```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.distplot(dataset,bins=25)
#combines histogram with kde
```

```python
sns.distplot(dataset,bins=25,rug=True,hist=False)
#turning some off
```

```python
sns.distplot(dataset,bins=25,
            kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue','label':'HIST'})
#can individually control the plots within the combine plot using dicts
```

```python
from pandas import Series
```

```python
ser1 = Series(dataset,name='new_data')
```

```python
sns.distplot(ser1,bins=25)
#seaborn works really well with pandas
```
