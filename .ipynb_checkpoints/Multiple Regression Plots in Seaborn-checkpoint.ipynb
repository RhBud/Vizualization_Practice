```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#tips vs total bill
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
#estimated linear fit model
```

```python
sns.lmplot('total_bill','tip',tips,scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#individually passing arguements
```

```python
sns.lmplot('total_bill','tip',tips,order=4,scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#now we're putting a different model on the data (not linear)
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
#basic scatter plot with no fit
```

```python
tips.head()
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
#create a tip percentage column
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
#size is x and tip percentage is y
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
#makes it a little easier to read
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
#shows the estimate for the tendency of each bin on how tip percentage depends on party size
#looks like more variance for party size of 1
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
#blue x's are female, o's are males. Shows the two regression plots
```

```python
#does the day make a difference in how much the tip is?
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
#whether tip percentage changed between days as well
```

```python
#can also do local regression model fitting in seaborn
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
#lmplot is actually using a lower level function which we can look at
```

```python
sns.regplot('total_bill','tip_pect',tips)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(tips['tip_pect'],tips['size'],color='red',ax=axis2)
```
