```python
#Monte Carlo Method
#Running simulations and using the aggregation of the results to analyze risk of the stock
```

```python
#geometric browninian motion where past information is independent of future actions
```

```python
#simulation of random drift and shock thousands of times to try and predict what the next stock movement will be 
```

```python
days = 365
dt = float(1)/365
mu = rets.mean()['GOOG']
#mu is associated for our drift
sigma = rets.std()['GOOG']
#sigma is volatility
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    
    price = np.zeros(days)
    #array of zeros the same length as days
    price[0] = start_price
    
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    #for every days we go along we grab a shock term grabbed from a normal ditribution
    #where scale is equal to sigma times sqrt of delta t
    
    for x in xrange(1,days):
        
        shock[x] = np.random.normal(loc=mu*dt,scale = sigma*np.sqrt(dt))
        
        drift[x] = mu * dt
        
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
        #current price therefore is equal to old price (at x-1) plus old price times drift and shock
```

```python
GOOG.head()
```

```python
#lets grab the oldest date
```

```python
start_price = 538.119995

for run in xrange(100):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
    #runs stock monte carlo function and it does this 100 times
    
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis for Google')
```

```python
#lets make a histogram of these larger results
```

```python

# Set a large numebr of runs
runs = 10000

# Create an empty matrix to hold the end price data
simulations = np.zeros(runs)

# Set the print options of numpy to only display 0-5 points from an array to suppress output
np.set_printoptions(threshold=5)

for run in xrange(runs):    
    # Set the simulation data point as the last stock price for that run
    simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1];
```

```python
q = np.percentile(simulations,1)
#uses numpy to define q as the 1 percent quartile
plt.hist(simulations,bins=200)

#chart all the titles and labels
# Starting Price
plt.figtext(0.6, 0.8, s="Start price: $%.2f" %start_price)
# Mean ending price
plt.figtext(0.6, 0.7, "Mean final price: $%.2f" % simulations.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,))

# Display 1% quantile
plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q)

# Plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# Title
plt.title(u"Final price distribution for Google Stock after %s days" % days, weight='bold');
```

```python
#red line at 1% quantile mark
#that means that the value at risk for each google stock is going to be 22 dollars 
#(that's the value at risk over a year)
#99% percent of the time, the most you could lose is $22.11
```
