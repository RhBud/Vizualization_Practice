```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset1 = randn(25)
```

```python
sns.rugplot(dataset1)
plt.ylim(0,1)
```

```python
plt.hist(dataset1,alpha=0.3)
sns.rugplot(dataset1)
```

```python
#kernel density represents each ticks marks with gaussian basis function
```

```python
sns.rugplot(dataset1)
x_min = dataset1.min() - 2
x_max = dataset1.max() + 2
#set two values for min and max to use in plotting

x_axis = np.linspace(x_min,x_max,100)
#give desired range of x axis and number of dilimeters based of created mins and max

#set bandwith using practical estimation equation
bandwidth = ((4*dataset1.std()**5)/(3*len(dataset1))) ** 0.2

kernel_list = []

#create a kernel for each point for every data point in dataset and append it to the kernel_list
for data_point in dataset1:
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel 
    
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list,axis=0)
fig = plt.plot(x_axis,sum_of_kde,color='indianred')
sns.rugplot(dataset1)
plt.yticks([])
plt.suptitle("Sum of the basis functions")
```

```python
sns.rugplot(dataset1,color='black')

for bw in np.arange(0.5,2.0,0.25):
    sns.kdeplot(dataset1,bw=bw,lw=1.8,label=bw)
```

```python
#kde plot run with various bandwidths
```

```python
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset1,kernel=kern,label=kern)
```

```python
#different estimation plots depending on what kernel type you use
```

```python
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset1,kernel=kern,label=kern,shade = True)
    #with shading
```

```python
#multivariate density estimation
```

```python
mean = [0,0]
cov = [[1,0],[0,100]]

dataset2 = np.random.multivariate_normal(mean,cov,1000)

dframe = pd.DataFrame(dataset2,columns=['X','Y'])
sns.kdeplot(dframe)
#using seaborn in syn with pandas dataframe
```

```python
#can also do this by passing in variables seperately and shade it
sns.kdeplot(dframe.X,dframe.Y,shade=True)
```

```python
#can specify bandwidths as well 
```

```python
sns.kdeplot(dframe,bw=1)
```

```python
sns.kdeplot(dframe,bw='silverman')
```

```python
#can also create a kde joint plot
```

```python
sns.jointplot('X','Y',dframe,kind='kde')
#same as other jointplots we did but specifying kde
```
